```python
import json
from datasets import Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer
from trl.trainer.sft_trainer import SFTTrainer
from trl.trainer.sft_config import SFTConfig


print("Loading base model and dataset")
model_name = "distilgpt2"

# Load Tokenizer and Model
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_name)

formatted_data = []

with open("restaurant_data.jsonl", "r") as f:
    for line in f:
        item = json.loads(line)
        item["text"] = f"Instruction: {item['instruction']}\nInput: {item['input']}\nOutput: {item['output']}"
        formatted_data.append(item)

dataset = Dataset.from_list(formatted_data)

training_args = SFTConfig(
    output_dir="./restaurant_model_results",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=2,
    num_train_epochs=5,
    learning_rate=2e-4,
    logging_steps=1,
    fp16=True,
    dataset_text_field="text",
)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    args=training_args,
)

print("Starting training on Colab GPU...")
trainer.train()

model.save_pretrained("./my_restaurant_ai")
tokenizer.save_pretrained("./my_restaurant_ai")
print("Training complete! Model saved to './my_restaurant_ai'.")
```

```python
from transformers import pipeline, AutoTokenizer


tokenizer = AutoTokenizer.from_pretrained("./my_restaurant_ai")


generator = pipeline(
    "text-generation",
    model="./my_restaurant_ai",
    tokenizer=tokenizer,
    device=0
)

test_name = "Luigi's Mansion"
test_cuisine = "Italian"
test_location = "Napoli"

prompt = f"Instruction: Write a short restaurant description.\nInput: Name: {test_name}, Cuisine: {test_cuisine}, Location: {test_location}\nOutput:"

print(f"\nGenerating description for {test_name}...\n")

result = generator(
    prompt,
    max_new_tokens=100,
    num_return_sequences=1,
    temperature=0.7,
    repetition_penalty=1.2,
    pad_token_id=tokenizer.eos_token_id
)

raw_output = result[0]['generated_text']
final_description = raw_output.split("Output:")[-1].strip()

print("Result")
print(final_description)
```
