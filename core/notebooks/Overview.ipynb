```python
import json
from datasets import Dataset
from transformers import AutoModelForCausalLM, AutoTokenizer
from trl.trainer.sft_trainer import SFTTrainer
from trl.trainer.sft_config import SFTConfig


print("Loading base model and dataset")
model_name = "distilgpt2"

# Load Tokenizer and Model
tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(model_name)

formatted_data = []

with open("restaurant_data.jsonl", "r") as f:
    for line in f:
        item = json.loads(line)
        item["text"] = f"Instruction: {item['instruction']}\nInput: {item['input']}\nOutput: {item['output']}"
        formatted_data.append(item)

dataset = Dataset.from_list(formatted_data)

training_args = SFTConfig(
    output_dir="./restaurant_model_results",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=2,
    num_train_epochs=5,
    learning_rate=2e-4,
    logging_steps=1,
    fp16=True,
    dataset_text_field="text",
)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    args=training_args,
)

print("Starting training on Colab GPU...")
trainer.train()

model.save_pretrained("./my_restaurant_ai")
tokenizer.save_pretrained("./my_restaurant_ai")
print("Training complete! Model saved to './my_restaurant_ai'.")
```
