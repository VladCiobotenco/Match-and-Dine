```python
import json
import torch
from datasets import Dataset
from google.colab import userdata

from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    BitsAndBytesConfig
)
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer, SFTConfig
from huggingface_hub import login
hf_token = userdata.get('TOKEN')
login(token = hf_token)

model_name = "google/gemma-2-2b-it"

bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map={"": 0},
    torch_dtype=torch.float16
)

model = prepare_model_for_kbit_training(model)

lora_config = LoraConfig(
    r=8,
    lora_alpha=16,
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)

formatted_data = []
with open("restaurant_data.jsonl", "r") as f:
    for line in f:
        item = json.loads(line)
        item["text"] = f"Instruction: {item['instruction']}\nInput: {item['input']}\nOutput: {item['output']}"
        formatted_data.append(item)

dataset = Dataset.from_list(formatted_data)

training_args = SFTConfig(
    output_dir="./restaurant_gemma_results",
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    num_train_epochs=2,
    learning_rate=5e-5,
    logging_steps=1,
    fp16=False,
    bf16=False,
    optim="paged_adamw_8bit",
    save_strategy="epoch",
    dataset_text_field="text",
    max_length=256
)

trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    args=training_args,
    processing_class=tokenizer
)

trainer.train()
trainer.save_model("./my_restaurant_ai")


repo_id = "Edy317/my_restaurant_aiv2"

model.push_to_hub(repo_id)
tokenizer.push_to_hub(repo_id)
print(f" Modelul a fost înlocuit pe https://huggingface.co/{repo_id}")
```

```python
from transformers import pipeline, AutoTokenizer


tokenizer = AutoTokenizer.from_pretrained("./my_restaurant_ai")


generator = pipeline(
    "text-generation",
    model="./my_restaurant_ai",
    tokenizer=tokenizer,
    device=0
)

test_name = "Edy's House"
test_cuisine = "Mexican"


prompt = f"Instruction: Write a short restaurant description.\nInput: Name: {test_name}, Details: {test_cuisine}\nOutput:"

print(f"\nGenerating description for {test_name}...\n")

result = generator(
    prompt,
    max_new_tokens=100,
    num_return_sequences=1,
    temperature=0.7,
    repetition_penalty=1.2,
    pad_token_id=tokenizer.eos_token_id
)

raw_output = result[0]['generated_text']
final_description = raw_output.split("Output:")[-1].strip()

print("Result")
print(final_description)
```
